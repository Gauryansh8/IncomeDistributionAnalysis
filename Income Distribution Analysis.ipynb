```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'income': [22000, 48000, 36000, 54000, 32000, 71000, 46000]}
df = pd.DataFrame(data)
df
```

```python
# Histogram
plt.hist(df['income'], bins=5, color='orange', edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
